```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()
kf = KFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_val_score(model, iris.data, iris.target, cv=kf)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
model = DecisionTreeClassifier(max_depth=2)
LOO =  LeaveOneOut()
# Use cross_val_score to evaluate the decision tree classifier using leave-one-out cross-validation
scores = cross_val_score(model, iris.data, iris.target, cv=LOO)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
