```python
from sklearn import datasets
iris_data = datasets.load_iris()
```

```python
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# HW6 วนลูปทำ 10-fold cross validation

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
# Define feature names
feature_names = iris_data.feature_names

# Define target names
target_names = iris_data.target_names

# Define Decision Tree Classifier
clf = DecisionTreeClassifier()

# Define KFold with 10 folds
kf = KFold(n_splits=10)

# Initialize lists to store results
accuracies = []
feature_importances = []

# Loop through each fold
for fold, (train_index, test_index) in enumerate(kf.split(iris_data.data)):
    
    # Get train and test data
    X_train, X_test = iris_data.data[train_index], iris_data.data[test_index]
    y_train, y_test = iris_data.target[train_index], iris_data.target[test_index]
    
    # Fit classifier on training data
    clf.fit(X_train, y_train)
    
    # Make predictions on test data
    y_pred = clf.predict(X_test)
    
    # Calculate accuracy score
    accuracy = accuracy_score(y_test, y_pred)
      # Store accuracy score in list
    accuracies.append(accuracy)
    
    # Get feature importances
    feature_importance = clf.feature_importances_
    
    # Store feature importances in list
    feature_importances.append(feature_importance)

    # Create a DataFrame to display the iris data for each fold
    iris_data1 = np.concatenate((X_test, y_test.reshape(-1,1)), axis=1)
    df = pd.DataFrame(data=iris_data1, columns=feature_names+['label'])
    df['fold'] = fold+1

    # Print fold number and accuracy score
    print(f"Fold {fold+1}")
    print("Cross-validation accuracy: %0.2f (+/- %0.2f)" % (accuracy, np.std(accuracies) * 2))
    print(df)
    print("\n")
```
